```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
```

```python
originalData = pd.read_csv('C:/Users/schar/Documents/PoliSciResearch/vars/purposecode_moneydistribution_survey.csv', engine="python")
df = originalData[['Purpose Code', 'Categorization']]
trainingSet = df.iloc[0:1001]
```

```python
#split by category
media = trainingSet[trainingSet["Categorization"] == "media"]
digital = trainingSet[trainingSet["Categorization"] == "digital"]
polling = trainingSet[trainingSet["Categorization"] == "polling"]
legal = trainingSet[trainingSet["Categorization"] == "legal"]
field = trainingSet[trainingSet["Categorization"] == "field"]
consulting = trainingSet[trainingSet["Categorization"] == "consulting"]
administrative = trainingSet[trainingSet["Categorization"] == "administrative"]
fundraising = trainingSet[trainingSet["Categorization"] == "fundraising"]
```

```python
#given a series, make it a whole corpus to analyze
stop_words = set(stopwords.words('english'))
def seriesToCorpus(df):
    phrases = list(df["Purpose Code"])
    corpus = " ".join(phrases)
    #remove stop words
    tokenized = word_tokenize(corpus)
    #tokenized = [w for w in tokenized if not w in stop_words]
    return " ".join(tokenized)

#plot top 20 words
def top(corpus, title):
    tok = nltk.tokenize.word_tokenize(corpus)
    bigrams = nltk.bigrams(tok)
    freqs = nltk.FreqDist(bigrams)
    freqs.plot(30, title = title, cumulative=False)

def corpusAndPlot(df, title):
    top(seriesToCorpus(df), title)
```

# Media

```python
corpusAndPlot(media, "Top Bigrams in Media")
```
